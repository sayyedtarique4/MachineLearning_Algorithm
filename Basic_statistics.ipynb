```python
# IQR method  
# Q1-1.5IQR   <  x   <Q3+1.5IQR
# Z-score method 
```

```python
Q1=np.quantile(lst,0.25)  # 25% of data
#Q2=np.quantile(lst,0.5)  # medianQ
Q3=np.quantile(lst,0.75)   # 75% of data
IQR=Q3-Q1  # Inter quartile range
print('Q1:',Q1,'Q3:',Q3,'IQR',IQR)
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
print('LowerBound :',lb,'Upper bound :',ub)
```

```python
outlier=[]
for i in lst:
    if (i<lb or i>ub):
        outlier.append(i)
```

```python
print(outlier)
```

```python
lst1=[]
for i in lst:
    if i not in outlier:
        lst1.append(i)
```

```python
print(lst1)
```

```python
sns.boxplot(lst1)
```

```python
# z-score method 
# zscore=x-xbar/std dev   xbar==mean
# -3 to 3
# point having z score <-3 and >3 they are treated as outliers
```

```python
import numpy as np
```

```python
lst=[1,2,3,3,4,4,5,6,8,10,6,7,9,10,24,15,25,22,100]
mean_lst=np.mean(lst)
std_dev=np.std(lst)
print(mean_lst,std_dev)
```

```python
threshold=3
outlier=[]
for i in lst:
    z_score=(i-mean_lst)/std_dev  # formula of z score
    abs_z=np.abs(z_score)
    if abs_z>3:
        outlier.append(i)
print(outlier)

```
